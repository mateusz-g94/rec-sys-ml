```python
def f(s):
    s2 = pd.Series(0, index=s.index)
    s2.iloc[-1] = 1
    return s2
```

```python
ratings['flag_last'] = ratings.sort_values(by=['userId', 'timestamp']).groupby('userId')['timestamp'].apply(f)
```
